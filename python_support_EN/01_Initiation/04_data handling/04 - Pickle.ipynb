Writing to a File
-----------------

```python
with open("test.pkl", "wb") as f:
    pickle.dump(data, f)
```

```python
with open("test.pkl", "rb") as f:
    content = pickle.load(f)

print(content, type(content))
```

Pickle can be used to save raw data to a file, but it is also used as a data exchange format over the network between two Python applications. Indeed, only bytes can be transmitted over the network.

Pickle evolves with the language itself, and during the migration to Python 3, a new protocol was created. Since then, it has been improved twice.
