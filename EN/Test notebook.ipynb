```python
def is_palindrome(mot: str) -> bool:
    """
    Dit si un mot est un palindrome

    >>> est_palindrome("Anna")
    True
    >>> est_palindrome("Viviane")
    False
    """
    return mot[::-1].lower() == mot.lower()
```

```python
from typing import Literal

def rochambeau(choice: Literal["rock", "paper", "scissors"]) -> str:
    """
    Playing Rock Paper Scissors
    >>> pierre_feuille_ciseaux("rock")
    you have won !
    >>> pierre_feuille_ciseaux("paper")
    you have lost !
    >>> pierre_feuille_ciseaux("scissors")
    this is a draw !
    """
    computer_choice = "scissors"
    universe = ["rock", "paper", "scissors"]

    result = (universe.index(choice) - universe.index(computer_choice)) % 3

    if result == 0:
        return "this is a draw !"
    if result == 1:
        return "you have won !"
    return "you have lost !"
```

```python
from datetime import date

def calculate_age(birthdate: date) -> int:
    """
    Calculate the age of a person from the date of birth

    >>> calculate_age(date(2000, 1, 1))
    24
    >>> calculate_age(date(2000, 12, 31))
    23
    """
    delta = date.today() - birthdate
    return delta.days // 365
```

```python
calculate_age(date(2009, 7, 11))
```

```python
from datetime import time, timedelta, datetime

def calculate_end(start: time, duration: int | float) -> time:
    """
    Calculate the end time, knowing the start time and the duration

    >>> calculate_end(time(10, 0, 0), 1)
    datetime.time(11, 0)
    >>> calculate_end(time(9, 30, 0), 2.5)
    datetime.time(12, 0)
    >>> calculate_end(time(23, 30, 0), 2)
    datetime.time(1, 30)
    """
    datetime_start = datetime.combine(date.today(), start)
    datetime_stop = datetime_start + timedelta(days=0, seconds=duration * 3600)
    return datetime_stop.time()
```

```python
calculate_end(time(10, 0, 0), 1)
```

```python
calculate_end(time(23, 30, 0), 2)
```

```python
from enum import unique, StrEnum

@unique
class Sex(StrEnum):
    M = "M"
    F = "F"

@unique
class Track(StrEnum):
    Finance = "Finance"
    Mkt = "Mkt"
    Cpta = "Cpta"
    RH = "RH"
```

```python
from typing import Annotated
from pydantic import AfterValidator

def is_a_grade(value: int) -> int:
    if 0 <= value <= 20:
        return value
    raise ValueError(f"Grade {value} should be between 0 and 20")

class Student(BaseModel):
    first_name: str
    last_name: str
    sex: Sex
    age: int
    track: Track
    grade: Annotated[int, AfterValidator(is_a_grade)]
```

```python
from pydantic import computed_field

class Student(BaseModel):
    first_name: str
    last_name: str
    sex: Sex
    age: int
    track: Track
    grade: int

    @computed_field
    @property
    def email(self) -> str:
        return f"{self.first_name}.{self.last_name}@edhec.com".lower()

    @computed_field
    @property
    def title(self) -> str:
        return "Mr" if self.sex == Sex.M else "Ms"

    @computed_field
    @property
    def points(self) -> int | None:
        result = 12 - self.grade
        return result if self.grade < 10 else None

    @field_validator("grade", mode="after")
    @classmethod
    def is_a_grade(cls, value: int) -> int:
        if 0 <= value <= 20:
            return value
        raise ValueError(f"Grade {value} should be between 0 and 20")
```
