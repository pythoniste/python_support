Decorators
==

```python
def decorateur(func):
    print("avant", func)
    return func
```

```python
@decorateur
def func():
    print("pendant")
```

```python
def func():
    print("pendant")
func = decorateur(func)
```

```python
func()
```

```python
def decorateur(func):
    def remplacement():
        print("avant", func)
        func()
        print("après", func)
    return remplacement

@decorateur
def func():
    print("pendant")

func()
```

```python
def handle(keys):
    def decorateur(transformation):
        def remplacement(data):
            for key in keys:
                data[key] = transformation(data[key])
        return remplacement
    return decorateur

@handle(["clé1", "clé3"])
def double(data):
    return data * 2

resultats = {
    "clé1": 1,
    "clé2": 2,
    "clé3": 3,
    "clé4": 4,
    "clé5": 5,
}

double(resultats)

print(resultats)
```

```python
@handle(["clé1", "clé4", "clé5"])
def triple(data):
    return data * 3

triple(resultats)
print(resultats)
```
