JSON
==

JSON is a data serialization technique that allows data to be exchanged over a network, potentially between heterogeneous programs.

```python
data = [
    {
        "pk": 1,
        "label": "test",
    }, {
        "pk": 2,
        "label": "python",        
    }
]
```

```python
serialized = '[{"pk": 1, "label": "test"}, {"pk": 2, "label": "python"}]'
```

---

Module Analysis
---------------

```python
import json
```

```python
type(json)
```

```python
dir(json)
```

Serializer
----------

```python
help(json.dump)
```

```python
with open("result.json", "w") as f:
    json.dump(data, f)
```

```python
with open("result.json") as f:
    print(f.read())
```

```python
help(json.dumps)
```

```python
json.dumps(data)  # Sert à sérialiser
```

Unserializer
------------

```python
help(json.loads)
```

```python
json.loads(serialized)  # Sert à désérialiser.
```

```python
obj = json.loads(serialized)
```

```python
from pprint import pprint
pprint(obj)
```

```python
type(obj)
```

```python
len(obj)
```

```python
help(json.load)
```

```python
with open("result.json") as f:
    result = json.load(f)
```

```python
print(result)
```

```python
type(result)
```

```python
len(result)
```
