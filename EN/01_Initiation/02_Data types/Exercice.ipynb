```python
from datetime import date, timedelta
incident = {
    "titre": "Panne routeur",
    "date_signalement": date.today(),
    "niveau": 1,
}
```

```python
liste = [
    incident,
    {
        "titre": "Perte de connexion",
        "date_signalement": date.today() - timedelta(days=1),
        "niveau": 1,
    },
    {
        "titre": "Application non accessible",
        "date_signalement": date.today() - timedelta(days=3),
        "niveau": 2,
    },
    {
        "titre": "Panne routeur",
        "date_signalement": date.today() - timedelta(days=3),
        "niveau": 2,
    },
    {
        "titre": "Application non accessible",
        "date_signalement": date.today() - timedelta(days=2),
        "niveau": 3,
    },
]
```

```python
list(liste)
```

```python
[incident for incident in liste if incident["niveau"] >= 2]
```

```python
pprint([incident for incident in liste if incident["date_signalement"] < date.today()])
```

```python
{incident["niveau"] for incident in liste if incident["date_signalement"] == date.today() - timedelta(days=3)}
```

```python
liste.remove(incident)
```
