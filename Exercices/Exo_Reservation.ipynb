Objects to create using **pydantic**:

- Person
  - id (UUID)
  - first name (str)
  - last name (str)
  - email (property firstname.lastname@mines.org)
  - date of birth (date)
  - age (property calculated from date of birth)
  - role (enum: professor or student)
- Room
  - number (int)
  - seating capacity (int)
- Booking
  - person
  - room
  - date_start (datetime)
  - date_end (datetime)
  - duration (property in hour + setter, changing it changes the end date)
  - status (enum: draft (default), confirmed, canceled)
