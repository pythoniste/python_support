TP Système et réseau : Logiciel de gestion de parc informatique
===============================================================

L'idée générale de l'exercice est de se donner des outils pour gérer un parc informatique.

On imagine que sur ce parc, on plein d'outils différents que l'on souhaite suivre et que certains d'entre eux peuvent être munis de serveurs TCP que l'on peut interroger à distance pour obtenir des informations sur leur état.

Partie 1
--------

1. Créer une fonction Python qui fait un appel système à la commande **df** pour récupérer le pourcentage d'espace libre sur **/** (le résultat est un float entre 0 et 1)
2. Créer une fonction Python qui fait un appel système à la commande **free** pour récupérer le pourcentage de mémoire libre sur l'espace mémoire et l'espace d'échange (le résultat est un 2-uplet de float entre 0 et 1)
3. Créer un serveur TCP qui :
    - lorsqu'il reçoit **proc** appelle la fonction python crée au 1 et renvoie le résultat
    - lorsqu'il recoit **mem** appelle la fonction python crée au 2 et renvoie le résultat
4. Créer un client qui peut envoyer soit "proc", soit "mem" au serveur et qui va afficher le résultat sous la forme d'un pourcentage ("24%", par exemple)

Partie 2
--------

Le type d'un matériel peut-être :

- Ordinateur
    - de bureau
    - portable
    - tablette
    - smartphone
- Serveur
  - Serveur dédié
  - Serveur VPS
  - Serveur Cloud
  - NAS
- Périphériques
  - Imprimante
  - Scanner
  - Caméra
  - Micro
  - Disque dur externe
- Équipement réseau
  - Switch
  - Routeur
  - Borne Wi-Fi
  - Firewal
  - Modem

1. Créer un Enum pour gérer ces types de matériel (niveau 2). Créer une propriété pour donner la catégorie d'un type (niveau 1).

Par exemple, je déclare un type **NAS**, j'ai une propriété du type qui me dit que c'est un **Serveur**.

2. Créer une entité Pydantic permettant de modéliser un matériel, dont voici les propriétés

- identifiant (UUID)
- titre (str)
- type (Enum)
- mise_en_service (date)
- dernière_mise_a_jour (datetime)

3. Rajouter deux attributs calculés:
  - une pour calculer l'age en mois du matériel
  - l'autre pour calculer en secondes la dernière mise à jour

4. Créer une entité pour gérer un inventaire:
  - numéro de l'inventaire
  - date de l'inventaire

5. Créer une entité pour gérer un élément d'inventaire
  - inventaire
  - matériel
  - emplacement (str)
  - adresse réseau du matériel
  - port ou l'outil est accessible
  - date_creation (datetime)

6. Rajouter un attribut à l'inventaire pour lister les éléments d'inventaires qui le concerne, ainsi qu'une propriété pour pointer vers l'élément d'inventaire le plus récent.

7. Ecrire plusieurs fonctions:
  - Créer un matériel
  - Créer un inventaire
  - Créer un élément d'inventaire

8. Créer une factory pour chacun de ces objets

https://docs.sqlalchemy.org/en/20/orm/mapping_styles.html#mapped-class-behavior

- identifiant (UUID)

- type (Enum)
- contexte (dict, stockée en JSON)

- numéro de l'inventaire
- date de l'inventaire

3. Puis un élément d'inventaire

- inventaire
- matériel

6. Utiliser les factory pour générer 25 matériels, un inventaire qui contient 4 élément d'inventaires. Les enregistrer dans la base de données.
