```python
from enum import Enum

class Sex(Enum):
    Homme = "H"
    Femme = "F"

class Track(Enum):
    FINANCE = "Finance"
    MARKETING = "Mkt"
    COMPTABILITÉ = "Cpta"
    RH = "RH"

class Etudiant(BaseModel):
    Nom: str
    Prénom: str
    Sexe: Sex
    Age: int
    Spé: Track
    Note: int

[Etudiant.model_validate(datum) for datum in data] + [Etudiant.model_validate(etudiant.model_dump())]
```

```python
class Etudiant(BaseModel):
    Nom: str
    Prénom: str
    Sexe: Sex
    Age: int
    Spé: Track
    Note: int

    @property
    def email(self):
        return f"{self.Prénom}.{self.Nom}@edhec.com"

etudiants = [Etudiant.model_validate(datum) for datum in data]
etudiants[0].email
```

```python
from pydantic import computed_field

class Etudiant(BaseModel):
    Nom: str
    Prénom: str
    Sexe: Sex
    Age: int
    Spé: Track
    Note: int

    @computed_field
    @property
    def email(self) -> str:
        return f"{self.Prénom}.{self.Nom}@edhec.com"

    @computed_field
    @property
    def Points(self) -> int:
        return 12 - self.Note

etudiants = [Etudiant.model_validate(datum) for datum in data]
print(etudiants[0].email)
print(etudiants[0].Points)
etudiants[0]
```
